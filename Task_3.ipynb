capacity and range.

```python
import pandas as pd
```

```python
data = pd.read_csv("FEV-data-Excel.xlsx - Auta elektryczne.csv")
data.head(10)
```

```python
ev_data_condition = data.copy()
```

Finding correlation index using Numpy       

```python
import numpy as np
```

```python
battery_cap = np.array(ev_data_condition["Battery capacity [kWh]"])
range_cap = np.array(ev_data_condition["Range (WLTP) [km]"])
```

```python
correlation_matrix = np.corrcoef(battery_cap, range_cap)

print(f"Correlation: {correlation_matrix[0,1]}")
```

The correlation value being 0.81 (i.e between 0.7-0.9) shows that there is strong positive relation between the variables.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set
sns.set(style="whitegrid")

# scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=ev_data_condition['Battery capacity [kWh]'], 
    y=ev_data_condition['Range (WLTP) [km]'], 
    color='blue', alpha=0.6
)
#reg plot
sns.regplot(
    x=ev_data_condition['Battery capacity [kWh]'], 
    y=ev_data_condition['Range (WLTP) [km]'], 
    scatter=False, color='red', line_kws={"linewidth": 2}
)

# title and labels
plt.title('Relationship Between Battery Capacity and Range', fontsize=14)
plt.xlabel('Battery Capacity [kWh]', fontsize=12)
plt.ylabel('Range (WLTP) [km]', fontsize=12)

# Show the plot
plt.show()
```

(b) Insights from the Visualization

1. Positive Correlation:

The scatter plot likely shows a positive trend (as battery capacity increases, the range also increases).
This makes sense since larger batteries store more energy, allowing EVs to travel further.

2. Trend Line Confirmation:

The red regression line helps visualize the general trend.
A strong linear trend suggests a strong correlation, whereas scattered points indicate a weaker relationship.

3. Outliers & Variability (reason):

Some cars may deviate from the trend, meaning other factors (like aerodynamics, weight, and efficiency) impact range.
Some high-capacity battery EVs may have unexpectedly low range, possibly due to high power consumption.
