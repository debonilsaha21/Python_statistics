Task 4: Build an EV recommendation class. The class should allow users to input their
budget, desired range, and battery capacity. The class should then return the top three EVs
matching their criteria. (8+8 Marks)

```python
import pandas as pd
```

```python
data = pd.read_csv("FEV-data-Excel.xlsx - Auta elektryczne.csv")
data.head(10)
```

```python
ev_data = data.copy()
```

```python
#creating the class
class EV_Recommendation:
    def __init__(self, data):
        self.data = data  # Store the DataFrame

    def budget_basis(self, budget):
        return self.data[self.data['Minimal price (gross) [PLN]'] < budget]\
            .sort_values(by='Minimal price (gross) [PLN]', ascending=False)\
            .head(3)

    def range_basis(self, range_value):
        return self.data[self.data['Range (WLTP) [km]'] > range_value]\
            .sort_values(by='Range (WLTP) [km]', ascending=False)\
            .head(3)

    def battery_cap_basis(self, battery_cap):
        return self.data[self.data['Battery capacity [kWh]'] > battery_cap]\
            .sort_values(by='Battery capacity [kWh]', ascending=False)\
            .head(3)

    def custom_basis(self, budget, range_value, battery_cap):
        return self.data[
            (self.data['Minimal price (gross) [PLN]'] < budget) &
            (self.data['Range (WLTP) [km]'] > range_value) &
            (self.data['Battery capacity [kWh]'] > battery_cap)
            ].sort_values(
            by=['Minimal price (gross) [PLN]', 'Range (WLTP) [km]', 'Battery capacity [kWh]'],
            ascending=[False, False, False]
            ).head(3)

        

            
```

```python
my_car = EV_Recommendation(ev_data)  #creating the object from class
```

```python
my_car.budget_basis(300000)            #using methods of the class created
```

```python
my_car.range_basis(420)
```

```python
my_car.battery_cap_basis(90)
```

```python
my_car.custom_basis(budget=380000, range_value= 350, battery_cap=80)
```
