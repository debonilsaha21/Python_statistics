```python
data = pd.read_csv("FEV-data-Excel.xlsx - Auta elektryczne.csv")
data.head(5)
```

of 400 km.

```python
data.info()
```

```python
ev_data = data.copy()
```

a) Your task is to filter out EVs that meet these criteria.(2 Marks)

```python
ev_data_condition = ev_data[(
    ev_data['Minimal price (gross) [PLN]']<350000
) & (
    ev_data['Range (WLTP) [km]']>400
)]
ev_data_condition.head()
```

b) Group them by the manufacturer (Make).(6 marks)

```python
ev_group_manf = ev_data_condition.groupby(['Make'])
ev_group_manf.head()
```

c) Calculate the average battery capacity for each manufacturer. (8 Marks)

```python
result = pd.pivot_table(
    ev_data_condition, 
    values='Battery capacity [kWh]', 
    index='Make', 
    aggfunc='mean'
)
result.sort_values(['Battery capacity [kWh]'], ascending= False)
```
